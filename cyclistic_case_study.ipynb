```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

## Trip Frequency and Volume Analysis

```python
# Objective: Determine how often each user type takes trips and total number of trips

# 1. Count total trips per usertype
trip_counts = df_combined.groupby('usertype')['trip_id'].count().reset_index()
trip_counts.columns = ['usertype', 'total_trips']

# 2. Calculate proportion of trips
total_trips = trip_counts['total_trips'].sum()
trip_counts['proportion'] = trip_counts['total_trips'] / total_trips * 100

# 3. Trip frequency over time (monthly trends)
df_combined['month'] = df_combined['start_time'].dt.to_period('M')
monthly_trips = df_combined.groupby(['usertype', 'month'])['trip_id'].count().reset_index()
monthly_trips.columns = ['usertype', 'month', 'trip_count']
monthly_trips['month'] = monthly_trips['month'].astype(str)  # For plotting
```

```python
# --- Visualizations ---
# Set seaborn style
sns.set(style="whitegrid")

# 1. Total Trips by Usertype
plt.figure(figsize=(8, 6))
sns.barplot(data=trip_counts, x='usertype', y='total_trips')
plt.title('Total Trips by User Type')
plt.xlabel('User Type')
plt.ylabel('Number of Trips')
for i, row in trip_counts.iterrows():
    plt.text(i, row['total_trips'], f"{row['proportion']:.1f}%", ha='center', va='bottom')
plt.tight_layout()
plt.savefig('total_trips_by_usertype.png')
plt.show()
plt.close()


# 2. Monthly Trip Trends
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_trips, x='month', y='trip_count', hue='usertype', marker='o')
plt.title('Monthly Trip Trends by User Type')
plt.xlabel('Month')
plt.ylabel('Number of Trips')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('monthly_trip_trends.png')
plt.show()
plt.close()
```

## Trip Duration Analysis

```python
# Objective: Compare the length of trips between user types

# 1. Calculate summary statistics for tripduration
# Convert to minutes for readability
df_combined['tripduration_min'] = df_combined['tripduration'] / 60
duration_stats = df_combined.groupby('usertype')['tripduration_min'].agg(['mean', 'median', 'min', 'max']).reset_index()

# 2. Categorize trip durations
bins = [0, 10, 30, float('inf')]
labels = ['Short (<10 min)', 'Medium (10-30 min)', 'Long (>30 min)']
df_combined['duration_category'] = pd.cut(df_combined['tripduration_min'], bins=bins, labels=labels, include_lowest=True)
duration_dist = df_combined.groupby(['usertype', 'duration_category'])['trip_id'].count().reset_index()
duration_dist.columns = ['usertype', 'duration_category', 'trip_count']

# 3. Trip duration by time of day and day of week
df_combined['hour'] = df_combined['start_time'].dt.hour
df_combined['day_of_week'] = df_combined['start_time'].dt.day_name()
duration_by_hour = df_combined.groupby(['usertype', 'hour'])['tripduration_min'].mean().reset_index()
duration_by_day = df_combined.groupby(['usertype', 'day_of_week'])['tripduration_min'].mean().reset_index()
# Order days of week for plotting
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
duration_by_day['day_of_week'] = pd.Categorical(duration_by_day['day_of_week'], categories=day_order, ordered=True)
duration_by_day = duration_by_day.sort_values('day_of_week')
```

```python
# --- Visualizations ---
# 3. Trip Duration Distribution
plt.figure(figsize=(10, 6))
sns.barplot(data=duration_dist, x='duration_category', y='trip_count', hue='usertype')
plt.title('Trip Duration Distribution by User Type')
plt.xlabel('Trip Duration Category')
plt.ylabel('Number of Trips')
plt.tight_layout()
plt.savefig('trip_duration_distribution.png')
plt.show()
plt.close()

# 4. Average Trip Duration by Hour
plt.figure(figsize=(12, 6))
sns.lineplot(data=duration_by_hour, x='hour', y='tripduration_min', hue='usertype', marker='o')
plt.title('Average Trip Duration by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Average Trip Duration (Minutes)')
plt.tight_layout()
plt.savefig('duration_by_hour.png')
plt.show()
plt.close()

# 5. Average Trip Duration by Day of Week
plt.figure(figsize=(12, 6))
sns.lineplot(data=duration_by_day, x='day_of_week', y='tripduration_min', hue='usertype', marker='o')
plt.title('Average Trip Duration by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Average Trip Duration (Minutes)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('duration_by_day.png')
plt.show()
plt.close()

# Save summary statistics to CSV for reference
trip_counts.to_csv('trip_counts.csv', index=False)
duration_stats.to_csv('duration_stats.csv', index=False)
```

## Time of Use (Temporal Patterns) Analysis

```python
# Objective: Identify when each user type uses bikes (time of day, day of week, season)

# 1. Extract time components
df_combined['hour'] = df_combined['start_time'].dt.hour
df_combined['day_of_week'] = df_combined['start_time'].dt.day_name()
df_combined['month'] = df_combined['start_time'].dt.to_period('M').astype(str)  # Convert to string for plotting

# 2. Group trips by usertype and time variables
# Hourly trip counts
hourly_trips = df_combined.groupby(['usertype', 'hour'])['trip_id'].count().reset_index()
hourly_trips.columns = ['usertype', 'hour', 'trip_count']

# Day of week trip counts
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df_combined['day_of_week'] = pd.Categorical(df_combined['day_of_week'], categories=day_order, ordered=True)
daily_trips = df_combined.groupby(['usertype', 'day_of_week'])['trip_id'].count().reset_index()
daily_trips.columns = ['usertype', 'day_of_week', 'trip_count']

# Monthly (seasonal) trip counts
monthly_trips = df_combined.groupby(['usertype', 'month'])['trip_id'].count().reset_index()
monthly_trips.columns = ['usertype', 'month', 'trip_count']
```

```python
# --- Visualizations ---
# Set seaborn style
sns.set(style="whitegrid")

# 1. Hourly Trip Counts
plt.figure(figsize=(12, 6))
sns.lineplot(data=hourly_trips, x='hour', y='trip_count', hue='usertype', marker='o')
plt.title('Trip Counts by Hour of Day and User Type')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Trips')
plt.xticks(range(0, 24))
plt.tight_layout()
plt.savefig('hourly_trip_counts.png')
plt.show()
plt.close()

# 2. Daily Trip Counts
plt.figure(figsize=(12, 6))
sns.barplot(data=daily_trips, x='day_of_week', y='trip_count', hue='usertype')
plt.title('Trip Counts by Day of Week and User Type')
plt.xlabel('Day of Week')
plt.ylabel('Number of Trips')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('daily_trip_counts.png')
plt.show()
plt.close()

# 3. Monthly Trip Counts
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_trips, x='month', y='trip_count', hue='usertype', marker='o')
plt.title('Trip Counts by Month and User Type')
plt.xlabel('Month')
plt.ylabel('Number of Trips')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('monthly_trip_counts.png')
plt.show()
plt.close()

# Save summary data to CSV for reference
hourly_trips.to_csv('hourly_trip_counts.csv', index=False)
daily_trips.to_csv('daily_trip_counts.csv', index=False)
monthly_trips.to_csv('monthly_trip_counts.csv', index=False)
```

## Station Usage

```python
# Objective: Understand where each user type starts and ends their trips

# 1. Top 10 start and end stations by usertype
# Start stations
start_stations = df_combined.groupby(['usertype', 'from_station_id', 'from_station_name'])['trip_id'].count().reset_index()
start_stations.columns = ['usertype', 'from_station_id', 'from_station_name', 'trip_count']
top_start_stations = start_stations.groupby('usertype').apply(lambda x: x.nlargest(10, 'trip_count')).reset_index(drop=True)

# End stations
end_stations = df_combined.groupby(['usertype', 'to_station_id', 'to_station_name'])['trip_id'].count().reset_index()
end_stations.columns = ['usertype', 'to_station_id', 'to_station_name', 'trip_count']
top_end_stations = end_stations.groupby('usertype').apply(lambda x: x.nlargest(10, 'trip_count')).reset_index(drop=True)

# 2. Station preferences (combine start and end for overall usage)
station_usage = df_combined.groupby(['usertype', 'from_station_name'])['trip_id'].count().reset_index(name='start_count')
station_usage_end = df_combined.groupby(['usertype', 'to_station_name'])['trip_id'].count().reset_index(name='end_count')
station_usage = station_usage.merge(station_usage_end, left_on=['usertype', 'from_station_name'],
                                   right_on=['usertype', 'to_station_name'], how='outer')
station_usage['from_station_name'] = station_usage['from_station_name'].fillna(station_usage['to_station_name'])
station_usage['total_count'] = station_usage[['start_count', 'end_count']].sum(axis=1, skipna=True)
top_stations_overall = station_usage.groupby('usertype').apply(lambda x: x.nlargest(10, 'total_count')).reset_index(drop=True)

# 3. Round trips vs. one-way trips
df_combined['is_round_trip'] = df_combined['from_station_id'] == df_combined['to_station_id']
round_trip_counts = df_combined.groupby(['usertype', 'is_round_trip'])['trip_id'].count().reset_index()
round_trip_counts.columns = ['usertype', 'is_round_trip', 'trip_count']
round_trip_counts['trip_type'] = round_trip_counts['is_round_trip'].map({True: 'Round Trip', False: 'One-Way'})
```

```python
# --- Visualizations ---
# Set seaborn style
sns.set(style="whitegrid")

# 1. Top 10 Start Stations
plt.figure(figsize=(12, 8))
sns.barplot(data=top_start_stations, x='trip_count', y='from_station_name', hue='usertype')
plt.title('Top 10 Start Stations by User Type')
plt.xlabel('Number of Trips')
plt.ylabel('Station Name')
plt.tight_layout()
plt.savefig('top_start_stations.png')
plt.show()
plt.close()

# 2. Top 10 End Stations
plt.figure(figsize=(12, 8))
sns.barplot(data=top_end_stations, x='trip_count', y='to_station_name', hue='usertype')
plt.title('Top 10 End Stations by User Type')
plt.xlabel('Number of Trips')
plt.ylabel('Station Name')
plt.tight_layout()
plt.savefig('top_end_stations.png')
plt.show()
plt.close()

# 3. Overall Station Usage (Start + End)
plt.figure(figsize=(12, 8))
sns.barplot(data=top_stations_overall, x='total_count', y='from_station_name', hue='usertype')
plt.title('Top 10 Stations (Start + End) by User Type')
plt.xlabel('Total Number of Trips')
plt.ylabel('Station Name')
plt.tight_layout()
plt.savefig('top_stations_overall.png')
plt.show()
plt.close()

# 4. Round Trips vs. One-Way Trips
plt.figure(figsize=(10, 6))
sns.barplot(data=round_trip_counts, x='trip_type', y='trip_count', hue='usertype')
plt.title('Round Trips vs. One-Way Trips by User Type')
plt.xlabel('Trip Type')
plt.ylabel('Number of Trips')
plt.tight_layout()
plt.savefig('round_vs_oneway_trips.png')
plt.show()
plt.close()
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Data for Subscribers
counts_subscriber = [
    df_combined.loc[(df_combined["usertype"] == "Subscriber") & (df_combined["gender"] == "Male")].shape[0],
    df_combined.loc[(df_combined["usertype"] == "Subscriber") & (df_combined["gender"] == "Female")].shape[0]
]

# Data for Customers
counts_customer = [
    df_combined.loc[(df_combined["usertype"] == "Customer") & (df_combined["gender"] == "Male")].shape[0],
    df_combined.loc[(df_combined["usertype"] == "Customer") & (df_combined["gender"] == "Female")].shape[0]
]

# Pie chart for Subscribers
axes[0].pie(counts_subscriber, labels=["Male", "Female"], autopct='%1.1f%%',
            startangle=90, colors=['#2986cc', '#c90076'])
axes[0].set_title("Gender Distribution by Subscribers")

# Pie chart for Customers
axes[1].pie(counts_customer, labels=["Male", "Female"], autopct='%1.1f%%',
            startangle=90, colors=['#2986cc', '#c90076'])
axes[1].set_title("Gender Distribution by Customers")

plt.tight_layout()
plt.savefig('gender_distribution.png')
plt.show()
```

### Age Group Percentage by User Type

```python
# Define age bins and labels
bins = [0, 25, 35, 45, 60, 100]
labels = ['18-25', '26-35', '36-45', '46-60', '60+']

# Create a new column for Age Groups
df_combined["age_group"] = pd.cut(df_combined["age"], bins=bins, labels=labels, right=False)
```

```python
import matplotlib.pyplot as plt

# Ensure correct filtering and value counting
age_counts_customer = df_combined.loc[df_combined["usertype"] == "Customer", "age_group"].value_counts().reindex(labels)
age_counts_subscriber = df_combined.loc[df_combined["usertype"] == "Subscriber", "age_group"].value_counts().reindex(labels)

# Create subplots
fig, axis = plt.subplots(1, 2, figsize=(14, 6))

# Plot pie chart for Customers
axis[0].pie(age_counts_customer, autopct='%1.1f%%', startangle=90,
            colors=['lightblue', 'orange', 'green', 'red', 'purple'],
            pctdistance=1.2, textprops={'fontsize': 10})
axis[0].set_title("Age Distribution of Customer Riders")

# Plot pie chart for Subscribers
axis[1].pie(age_counts_subscriber, autopct='%1.1f%%', startangle=90,
            colors=['lightblue', 'orange', 'green', 'red', 'purple'],
            pctdistance=1.2, textprops={'fontsize': 10})
axis[1].set_title("Age Distribution of Subscriber Riders")

# Add a legend for both charts
fig.legend(labels=age_counts_customer.index, loc='center right', bbox_to_anchor=(1.15, 0.5))

# Adjust layout to prevent overlap
plt.tight_layout()
plt.savefig('age_distribution.png')
plt.show()
```
